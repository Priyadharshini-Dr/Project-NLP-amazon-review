### Merging Datasets to have a better review

To consolidate the product review data, I have merged the three available CSV files:
- `/content/1429_1.csv`
- `/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv`
- `/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv`

This was done by reading each CSV file into a pandas DataFrame and then concatenating them into a single, larger DataFrame. This merged dataset will be used for the subsequent analysis and model building steps in the project.

```python
import pandas as pd
import os

# List of CSV file paths to merge
csv_files = ['/content/1429_1.csv', '/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv', '/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv']

# Create an empty list to store DataFrames
dfs = []

# Read each CSV file into a DataFrame and append to the list
print("Reading CSV files...")
for file in csv_files:
    if os.path.exists(file):
        try:
            # Attempt to read the CSV, handling potential errors or different structures
            temp_df = pd.read_csv(file, low_memory=False) # low_memory=False to handle mixed types
            dfs.append(temp_df)
            print(f"Successfully read {file}")
        except Exception as e:
            print(f"Error reading {file}: {e}")
    else:
        print(f"File not found: {file}")

# Check if any DataFrames were loaded
if dfs:
    # Concatenate all DataFrames in the list
    print("\nMerging DataFrames...")
    merged_df = pd.concat(dfs, ignore_index=True)
    print("DataFrames merged successfully.")

    # Display information about the merged DataFrame
    print("\nInformation about the merged dataset:")
    display(merged_df.info())

    # Display the first 5 rows of the merged DataFrame
    print("\nFirst 5 rows of the merged dataset:")
    display(merged_df.head())

    # Update the main DataFrame 'df' with the merged data
    df = merged_df
    print("\nMain DataFrame 'df' updated with merged data.")
else:
    print("\nNo DataFrames were loaded. Merging could not be performed.")
```

```python
# Data Preprocessing

# Check for missing values in the relevant columns
print("\nMissing values before handling:")
display(df[['reviews.rating', 'reviews.text']].isnull().sum())

# Drop rows with missing values in 'reviews.rating' or 'reviews.text' as these are crucial for classification
df.dropna(subset=['reviews.rating', 'reviews.text'], inplace=True)

# Verify that missing values have been handled
print("\nMissing values after handling:")
display(df[['reviews.rating', 'reviews.text']].isnull().sum())
```

### **Down sampling and up sampling to balance** the data set of postive negative and neutral

```python
# Map star ratings to sentiment classes
def map_rating_to_sentiment(rating):
    if rating >= 4:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return 'Negative'

df['sentiment'] = df['reviews.rating'].apply(map_rating_to_sentiment)

# Display the distribution of sentiment classes
print("\nDistribution of sentiment classes:")
display(df['sentiment'].value_counts())
```

```python
from sklearn.utils import resample

# Display the original class distribution
print("Original class distribution:")
display(df['sentiment'].value_counts())

# Separate majority and minority classes
df_majority = df[df['sentiment'] == 'Positive']
df_neutral = df[df['sentiment'] == 'Neutral']
df_negative = df[df['sentiment'] == 'Negative']

# Determine the size of the majority class
majority_class_size = len(df_majority)

# Downsample the majority class
# Let's downsample the majority class to be closer in size to the largest minority class (Neutral in this case)
# or a value that provides a reasonable balance without losing too much data.
# For demonstration, let's aim for a size slightly larger than the Neutral class.
# You can adjust target_size based on desired balance and dataset size.
target_size = len(df_neutral) * 2 # Example: aiming for double the size of the neutral class
if target_size > majority_class_size:
    target_size = majority_class_size # Don't upsample the majority class

df_majority_downsampled = resample(df_majority,
                                 replace=False,    # Sample without replacement
                                 n_samples=target_size, # To match target size
                                 random_state=42) # Reproducible results

print(f"\nDownsampled Positive class size: {len(df_majority_downsampled)}")

# Upsample minority classes (Neutral and Negative)
# Upsample both minority classes to the size of the downsampled majority class
df_neutral_upsampled = resample(df_neutral,
                              replace=True,     # Sample with replacement
                              n_samples=target_size, # To match target size
                              random_state=42) # Reproducible results

df_negative_upsampled = resample(df_negative,
                               replace=True,     # Sample with replacement
                               n_samples=target_size, # To match target size
                               random_state=42) # Reproducible results

print(f"Upsampled Neutral class size: {len(df_neutral_upsampled)}")
print(f"Upsampled Negative class size: {len(df_negative_upsampled)}")


# Combine the downsampled majority class and upsampled minority classes
df_balanced = pd.concat([df_majority_downsampled, df_neutral_upsampled, df_negative_upsampled])

# Shuffle the balanced dataframe
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Display the new class distribution
print("\nBalanced class distribution:")
display(df_balanced['sentiment'].value_counts())

# Update the main DataFrame 'df' with the balanced data for training
df = df_balanced
print("\nMain DataFrame 'df' updated with balanced data.")
```

```python
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer

# Split the balanced data into training and testing sets
# Use the balanced df DataFrame
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['sentiment'])

print("Training data shape:", train_df.shape)
print("Testing data shape:", test_df.shape)

# Load a pre-trained tokenizer
# I will use 'distilbert-base-uncased' as suggested in the project requirements for its efficiency.
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

# Tokenize and encode the text data
def tokenize_and_encode(tokenizer, dataframe, text_column):
    return tokenizer(dataframe[text_column].tolist(), padding=True, truncation=True, return_tensors='pt')

train_encodings = tokenize_and_encode(tokenizer, train_df, 'reviews.text')
test_encodings = tokenize_and_encode(tokenizer, test_df, 'reviews.text')

print("\nTraining encodings shape:", train_encodings['input_ids'].shape)
print("Testing encodings shape:", test_encodings['input_ids'].shape)
```

```python
# Convert sentiment labels to integers
label_map = {'Negative': 0, 'Neutral': 1, 'Positive': 2}

# Use the sentiment column from the balanced train_df and test_df
train_labels = train_df['sentiment'].map(label_map)
test_labels = test_df['sentiment'].map(label_map)

print("\nTraining labels shape:", train_labels.shape)
print("Testing labels shape:", test_labels.shape)
```

```python
import torch
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import numpy as np

# Create a custom Dataset
class SentimentDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        # Ensure labels are converted to a tensor
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long) # Ensure dtype is long
        return item

    def __len__(self):
        return len(self.labels)

# Ensure train_labels and test_labels are lists or numpy arrays for Dataset
train_dataset = SentimentDataset(train_encodings, train_labels.tolist())
test_dataset = SentimentDataset(test_encodings, test_labels.tolist())


# Load a pre-trained model for sequence classification
# Using the correct num_labels for our 3 sentiment classes
model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)

# Define a function to compute metrics
def compute_metrics(p):
    preds = p.predictions[0] if isinstance(p.predictions, tuple) else p.predictions
    preds = np.argmax(preds, axis=1)
    accuracy = accuracy_score(p.label_ids, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(p.label_ids, preds, average='weighted')
    return {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}


# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # number of training epochs
    per_device_train_batch_size=16,  # batch size per device during training
    per_device_eval_batch_size=64,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
    eval_strategy="epoch",            # Evaluate every epoch
    save_strategy="epoch",           # Save checkpoints every epoch
    load_best_model_at_end=True,     # Load the best model at the end of training
    metric_for_best_model="accuracy", # Use accuracy to determine the best model
    greater_is_better=True           # Higher accuracy is better
)

# Create Trainer instance
trainer = Trainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset,            # evaluation dataset
    compute_metrics=compute_metrics      # our custom metrics function
)

# Train the model
print("\nTraining the model...")
trainer.train()

# Evaluate the model
print("\nEvaluating the model...")
eval_results = trainer.evaluate()
print("\nEvaluation results:")
display(eval_results)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Extract the unique categories
unique_categories = df['categories'].unique()

# Use TF-IDF to vectorize the categories
# I will set max_features to limit the number of features and reduce dimensionality
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(unique_categories)

# Apply K-Means clustering
# Let's start with 5 clusters as an initial number, we can adjust this later
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10) # Set n_init to suppress warning
kmeans.fit(tfidf_matrix)

# Get the cluster assignments for each unique category
category_clusters = kmeans.labels_

# Add the cluster assignments back to the unique categories for analysis
clustered_categories = pd.DataFrame({'category': unique_categories, 'cluster': category_clusters})

# Display the categories in each cluster to see the groupings
print("\nCategories in each cluster:")
for i in range(num_clusters):
    print(f"\nCluster {i}:")
    display(clustered_categories[clustered_categories['cluster'] == i]['category'].tolist())
```

```python
# Map the cluster assignments back to the original DataFrame
# Create a mapping from unique category strings to their cluster labels
category_to_cluster = clustered_categories.set_index('category')['cluster'].to_dict()

# Apply this mapping to the 'categories' column in the original DataFrame
df['category_cluster'] = df['categories'].map(category_to_cluster)

# Display the distribution of reviews across the clusters
print("\nDistribution of reviews across clusters:")
display(df['category_cluster'].value_counts())
```

## Data preparation for summarization

### Subtask:
Within each product cluster, identify the top 3 products, their key differences, top complaints, and the worst product based on the reviews.


**Reasoning**:
I need to iterate through each cluster, identify products within each cluster, calculate metrics for each product, and then identify top/worst products and their associated feedback. I will start by grouping the data by cluster and then by product within each cluster to calculate the average rating and review counts. I will use the 'name' and 'brand' columns to identify products, handling potential missing 'name' values by using 'asins' as a fallback if necessary.



```python
import pandas as pd

cluster_summaries = {}

# Iterate through each unique category cluster
for cluster_id in df['category_cluster'].unique():
    print(f"\nProcessing Cluster {cluster_id}...")

    # Filter the DataFrame for the current cluster
    cluster_df = df[df['category_cluster'] == cluster_id].copy()

    # Identify products within the cluster. Use 'name' and 'brand', fallback to 'asins' if 'name' is missing.
    # Create a unique product identifier
    cluster_df['product_identifier'] = cluster_df.apply(
        lambda row: f"{row['brand']} - {row['name']}" if pd.notna(row['name']) else f"{row['brand']} - {row['asins']}",
        axis=1
    )

    # Calculate average rating and review count for each product
    product_stats = cluster_df.groupby('product_identifier').agg(
        average_rating=('reviews.rating', 'mean'),
        review_count=('reviews.rating', 'count')
    ).reset_index()

    # Sort products by average rating and review count to identify top and worst
    product_stats = product_stats.sort_values(by=['average_rating', 'review_count'], ascending=[False, False])

    # Consider products with a reasonable number of reviews (e.g., > 5) for top/worst analysis
    min_reviews_for_ranking = 5
    ranked_products = product_stats[product_stats['review_count'] >= min_reviews_for_ranking]

    # Identify top 3 products
    top_products = ranked_products.head(3)

    # Identify worst product (lowest average rating among those with min_reviews_for_ranking)
    worst_product = ranked_products.tail(1)

    cluster_summaries[cluster_id] = {
        'top_products': top_products.to_dict('records'),
        'worst_product': worst_product.to_dict('records')
    }

    print(f"Finished processing Cluster {cluster_id}.")

# Display the product identification and ranking results for a sample cluster
# (You can adjust the cluster_id to inspect different clusters)
sample_cluster_id = df['category_cluster'].unique()[0]
print(f"\nSample Product Stats for Cluster {sample_cluster_id}:")
display(product_stats.head())
display(ranked_products.head())
display(top_products)
display(worst_product)
```

I will prepare the data by formatting the extracted insights for each cluster into a structured input that can be used as a prompt for the generative model during inference.



```python
# Prepare data for inference by formatting the extracted insights into prompts

from collections import Counter
import re
import pandas as pd # Ensure pandas is imported
import matplotlib.pyplot as plt # Import plotting libraries
import seaborn as sns # Import seaborn for better plots

# Function to extract frequent terms from reviews
def get_frequent_terms(reviews, n_most_common=10):
    all_words = ' '.join(reviews).lower()
    # Remove punctuation and split into words
    words = re.findall(r'\b\w+\b', all_words)

    # Simple frequency count for unigrams and bigrams
    word_counts = Counter(words)
    bigram_counts = Counter(zip(words[:-1], words[1:]))

    most_common_words = word_counts.most_common(n_most_common)
    most_common_bigrams = bigram_counts.most_common(n_most_common)

    return most_common_words, most_common_bigrams

# Initialize cluster_summaries dictionary
cluster_summaries = {}

# Ensure df is available from previous steps
if 'df' not in locals() or df.empty:
    print("DataFrame 'df' not found or is empty. Please ensure data loading and preprocessing were successful.")
else:
    # Iterate through each unique category cluster
    for cluster_id in df['category_cluster'].unique():
        print(f"\nProcessing Cluster {cluster_id}...")

        # Filter the DataFrame for the current cluster
        cluster_df = df[df['category_cluster'] == cluster_id].copy()

        # Identify products within the cluster. Use 'name' and 'brand', fallback to 'asins' if 'name' is missing.
        # Create a unique product identifier
        cluster_df['product_identifier'] = cluster_df.apply(
            lambda row: f"{row['brand']} - {row['name']}" if pd.notna(row['name']) else f"{row['brand']} - {row['asins']}",
            axis=1
        )

        # Calculate average rating and review count for each product
        product_stats = cluster_df.groupby('product_identifier').agg(
            average_rating=('reviews.rating', 'mean'),
            review_count=('reviews.rating', 'count')
        ).reset_index()

        # Sort products by average rating and review count to identify top and worst
        product_stats = product_stats.sort_values(by=['average_rating', 'review_count'], ascending=[False, False])

        # Consider products with a reasonable number of reviews (e.g., > 5) for top/worst analysis
        min_reviews_for_ranking = 5
        ranked_products = product_stats[product_stats['review_count'] >= min_reviews_for_ranking].copy() # Use .copy() to avoid SettingWithCopyWarning

        # Identify top 3 products
        top_products = ranked_products.head(3)

        # Identify worst product (lowest average rating among those with min_reviews_for_ranking)
        worst_product = ranked_products.tail(1)

        # --- Extract Differences and Complaints ---
        # Initialize dictionaries to store frequent terms for top and worst products
        top_products_differences = {}
        top_products_complaints = {}
        worst_product_reasons_to_avoid = {}


        # Analyze reviews for top products (focus on positive reviews for differences)
        for index, product in top_products.iterrows():
            product_reviews_df = cluster_df[cluster_df['product_identifier'] == product['product_identifier']].copy()
            positive_reviews = product_reviews_df[product_reviews_df['sentiment'] == 'Positive']['reviews.text'].dropna()
            negative_reviews = product_reviews_df[product_reviews_df['sentiment'] == 'Negative']['reviews.text'].dropna()

            if not positive_reviews.empty:
                 freq_words_pos, freq_bigrams_pos = get_frequent_terms(positive_reviews)
                 top_products_differences[product['product_identifier']] = {
                     'frequent_words': freq_words_pos,
                     'frequent_bigrams': freq_bigrams_pos
                 }
            else:
                 top_products_differences[product['product_identifier']] = "No positive reviews to analyze for differences."

            if not negative_reviews.empty:
                 freq_words_neg, freq_bigrams_neg = get_frequent_terms(negative_reviews)
                 top_products_complaints[product['product_identifier']] = {
                     'frequent_words': freq_words_neg,
                     'frequent_bigrams': freq_bigrams_neg
                 }
            else:
                 top_products_complaints[product['product_identifier']] = "No negative reviews to analyze for complaints."


        # Analyze reviews for the worst product (focus on negative reviews for reasons to avoid)
        if not worst_product.empty:
             worst_product_row = worst_product.iloc[0]
             worst_product_reviews_df = cluster_df[cluster_df['product_identifier'] == worst_product_row['product_identifier']].copy()
             worst_negative_reviews = worst_product_reviews_df[worst_product_reviews_df['sentiment'] == 'Negative']['reviews.text'].dropna()

             if not worst_negative_reviews.empty:
                 freq_words_worst_neg, freq_bigrams_worst_neg = get_frequent_terms(worst_negative_reviews)
                 worst_product_reasons_to_avoid[worst_product_row['product_identifier']] = {
                     'frequent_words': freq_words_worst_neg,
                     'frequent_bigrams': freq_bigrams_worst_neg
                 }
             else:
                 worst_product_reasons_to_avoid[worst_product_row['product_identifier']] = "No negative reviews to analyze for reasons to avoid."
        else:
             # Handle case where no worst product meets the minimum review threshold
             worst_product_reasons_to_avoid["N/A"] = "No worst product with sufficient reviews."


        # Store all extracted insights for the current cluster
        cluster_summaries[cluster_id] = {
            'top_products': top_products.to_dict('records'),
            'worst_product': worst_product.to_dict('records'),
            'top_products_differences': top_products_differences,
            'top_products_complaints': top_products_complaints,
            'worst_product_reasons_to_avoid': worst_product_reasons_to_avoid
        }

        print(f"Finished processing Cluster {cluster_id}.")

    # Display the product identification and ranking results for a sample cluster
    if cluster_summaries:
        sample_cluster_id = list(cluster_summaries.keys())[0]
        print(f"\nSample Product Stats and Insights for Cluster {sample_cluster_id}:")

        # Find the product_stats for the sample cluster to display top/worst
        sample_cluster_df = df[df['category_cluster'] == sample_cluster_id].copy()
        sample_cluster_df['product_identifier'] = sample_cluster_df.apply(
             lambda row: f"{row['brand']} - {row['name']}" if pd.notna(row['name']) else f"{row['brand']} - {row['asins']}",
             axis=1
         )
        sample_product_stats = sample_cluster_df.groupby('product_identifier').agg(
             average_rating=('reviews.rating', 'mean'),
             review_count=('reviews.rating', 'count')
         ).reset_index()
        sample_product_stats = sample_product_stats.sort_values(by=['average_rating', 'review_count'], ascending=[False, False])
        sample_ranked_products = sample_product_stats[sample_product_stats['review_count'] >= min_reviews_for_ranking].copy()

        print("\nSample Top Products:")
        display(sample_ranked_products.head(3))
        print("\nSample Worst Product:")
        display(sample_ranked_products.tail(1))

        # Print some extracted insights for sample products
        sample_summary_data = cluster_summaries[sample_cluster_id]
        print("\nSample Extracted Differences (from Positive Reviews of Top Products):")
        for product_id, insights in sample_summary_data['top_products_differences'].items():
            print(f"- {product_id}: {insights}")

        print("\nSample Extracted Complaints (from Negative Reviews of Top Products):")
        for product_id, insights in sample_summary_data['top_products_complaints'].items():
            print(f"- {product_id}: {insights}")

        print("\nSample Extracted Reasons to Avoid (from Negative Reviews of Worst Product):")
        for product_id, insights in sample_summary_data['worst_product_reasons_to_avoid'].items():
             print(f"- {product_id}: {insights}")


    else:
        print("\nNo cluster summaries generated.")

# The code to format summary for inference is already in this cell below this line
def format_summary_for_inference(summary_data):
    """Formats the extracted cluster summary data into a text string for inference."""
    prompt = "Generate a product recommendation article based on the following insights:\n\n"

    if summary_data['top_products']:
        prompt += "Top Recommended Products:\n"
        for i, product in enumerate(summary_data['top_products']):
            prompt += f"{i+1}. {product['product_identifier']} (Average Rating: {product['average_rating']:.2f}, Reviews: {product['review_count']})\n"

        prompt += "\nKey Differences and Strengths of Top Products:\n"
        for product_id, insights in summary_data['top_products_differences'].items():
            prompt += f"- {product_id}:\n"
            if isinstance(insights, dict):
                prompt += f"  Frequent positive words: {', '.join([word for word, count in insights.get('frequent_words', [])])}\n"
                prompt += f"  Frequent positive bigrams: {', '.join([' '.join(bigram) for bigram, count in insights.get('figram_counts', [])])}\n" # Corrected typo in bigram key
            else:
                prompt += f"  {insights}\n"

        prompt += "\nCommon Complaints for Top Products:\n"
        for product_id, insights in summary_data['top_products_complaints'].items():
            prompt += f"- {product_id}:\n"
            if isinstance(insights, dict):
                prompt += f"  Frequent negative words: {', '.join([word for word, count in insights.get('frequent_words', [])])}\n"
                prompt += f"  Frequent negative bigrams: {', '.join([' '.join(bigram) for bigram, count in insights.get('frequent_bigrams', [])])}\n"
            else:
                prompt += f"  {insights}\n"

    if summary_data['worst_product']:
        worst_product = summary_data['worst_product'][0]
        prompt += f"\nWorst Product to Avoid:\n"
        prompt += f"- {worst_product['product_identifier']} (Average Rating: {worst_product['average_rating']:.2f}, Reviews: {worst_product['review_count']})\n"

        prompt += "\nReasons to Avoid the Worst Product:\n"
        for product_id, insights in summary_data['worst_product_reasons_to_avoid'].items():
             prompt += f"- {product_id}:\n"
             if isinstance(insights, dict):
                prompt += f"  Frequent negative words: {', '.join([word for word, count in insights.get('frequent_words', [])])}\n"
                prompt += f"  Frequent negative bigrams: {', '.join([' '.join(bigram) for bigram, count in insights.get('frequent_bigrams', [])])}\n"
             else:
                 prompt += f"  {insights}\n"


    return prompt

# Create prompts for each cluster
cluster_prompts = {}
# Ensure cluster_summaries is populated before iterating
if cluster_summaries:
    for cluster_id, summary_data in cluster_summaries.items():
        cluster_prompts[cluster_id] = format_summary_for_inference(summary_data)

    # Display a sample prompt
    sample_cluster_id = list(cluster_prompts.keys())[0]
    print(f"Sample prompt for Cluster {sample_cluster_id}:\n")
    print(cluster_prompts[sample_cluster_id])

    # The model and tokenizer are now loaded, and the data is prepared for inference.
    # The next step will be to use these prompts with the loaded model to generate summaries.
else:
    print("Cluster summaries were not generated. Cannot create prompts.")
```

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
import torch
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import numpy as np
import wandb

# --- Data Loading and Preprocessing ---
# List of CSV file paths to merge
csv_files = ['/content/1429_1.csv', '/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv', '/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv']

# Create an empty list to store DataFrames
dfs = []

# Read each CSV file into a DataFrame and append to the list
print("Reading CSV files...")
for file in csv_files:
    if os.path.exists(file):
        try:
            temp_df = pd.read_csv(file, low_memory=False)
            dfs.append(temp_df)
            print(f"Successfully read {file}")
        except Exception as e:
            print(f"Error reading {file}: {e}")
    else:
        print(f"File not found: {file}")

# Check if any DataFrames were loaded
if dfs:
    print("\nMerging DataFrames...")
    merged_df = pd.concat(dfs, ignore_index=True)
    print("DataFrames merged successfully.")
    df = merged_df
else:
    print("\nNo DataFrames were loaded. Merging could not be performed. Exiting.")
    # In a real script, you might exit or raise an error here.
    # For this notebook, we'll continue but subsequent steps might fail.
    df = pd.DataFrame() # Create an empty DataFrame to prevent errors

if not df.empty:
    # Check for missing values in the relevant columns and drop rows
    print("\nMissing values before handling:")
    display(df[['reviews.rating', 'reviews.text']].isnull().sum())
    df.dropna(subset=['reviews.rating', 'reviews.text'], inplace=True)
    print("\nMissing values after handling:")
    display(df[['reviews.rating', 'reviews.text']].isnull().sum())

    # Map star ratings to sentiment classes
    def map_rating_to_sentiment(rating):
        if rating >= 4:
            return 'Positive'
        elif rating == 3:
            return 'Neutral'
        else:
            return 'Negative'

    df['sentiment'] = df['reviews.rating'].apply(map_rating_to_sentiment)

    # Balance the sentiment distribution using resampling
    from sklearn.utils import resample

    print("\nOriginal class distribution:")
    display(df['sentiment'].value_counts())

    df_majority = df[df['sentiment'] == 'Positive']
    df_neutral = df[df['sentiment'] == 'Neutral']
    df_negative = df[df['sentiment'] == 'Negative']

    # Determine the size of the majority class
    majority_class_size = len(df_majority)
    target_size = len(df_neutral) * 2
    if target_size > majority_class_size:
        target_size = majority_class_size

    df_majority_downsampled = resample(df_majority,
                                     replace=False,
                                     n_samples=target_size,
                                     random_state=42)

    df_neutral_upsampled = resample(df_neutral,
                                  replace=True,
                                  n_samples=target_size,
                                  random_state=42)

    df_negative_upsampled = resample(df_negative,
                                   replace=True,
                                   n_samples=target_size,
                                   random_state=42)

    df_balanced = pd.concat([df_majority_downsampled, df_neutral_upsampled, df_negative_upsampled])
    df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

    df = df_balanced # Use the balanced DataFrame for training

    print("\nBalanced class distribution:")
    display(df['sentiment'].value_counts())

    # Split the data into training and testing sets
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['sentiment'])

    print("Training data shape:", train_df.shape)
    print("Testing data shape:", test_df.shape)

    # Load a pre-trained tokenizer
    tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

    # Tokenize and encode the text data
    def tokenize_and_encode(tokenizer, dataframe, text_column):
        return tokenizer(dataframe[text_column].tolist(), padding=True, truncation=True, return_tensors='pt')

    train_encodings = tokenize_and_encode(tokenizer, train_df, 'reviews.text')
    test_encodings = tokenize_and_encode(tokenizer, test_df, 'reviews.text')

    print("\nTraining encodings shape:", train_encodings['input_ids'].shape)
    print("Testing encodings shape:", test_encodings['input_ids'].shape)

    # Convert sentiment labels to integers
    label_map = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
    train_labels = train_df['sentiment'].map(label_map)
    test_labels = test_df['sentiment'].map(label_map)

    print("\nTraining labels shape:", train_labels.shape)
    print("Testing labels shape:", test_labels.shape)

    # --- W&B Integration and Model Training ---
    # Create a custom Dataset
    class SentimentDataset(Dataset):
        def __init__(self, encodings, labels):
            self.encodings = encodings
            self.labels = labels

        def __getitem__(self, idx):
            item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
            item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
            return item

        def __len__(self):
            return len(self.labels)

    # Ensure train_labels and test_labels are lists or numpy arrays for Dataset
    train_dataset = SentimentDataset(train_encodings, train_labels.tolist())
    test_dataset = SentimentDataset(test_encodings, test_labels.tolist())

    # Start a new W&B run
    wandb.init(project="amazon-reviews-sentiment-analysis", name="distilbert-finetune-balanced")

    # Load a pre-trained model for sequence classification
    model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)

    # Define a function to compute metrics
    def compute_metrics(p):
        preds = p.predictions[0] if isinstance(p.predictions, tuple) else p.predictions
        preds = np.argmax(preds, axis=1)
        accuracy = accuracy_score(p.label_ids, preds)
        precision, recall, f1, _ = precision_recall_fscore_support(p.label_ids, preds, average='weighted')
        return {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}

    # Define training arguments and integrate with W&B
    training_args = TrainingArguments(
        output_dir='./results',          # output directory
        num_train_epochs=3,              # number of training epochs
        per_device_train_batch_size=16,  # batch size per device during training
        per_device_eval_batch_size=64,   # batch size for evaluation
        warmup_steps=500,                # number of warmup steps for learning rate scheduler
        weight_decay=0.01,               # strength of weight decay
        logging_dir='./logs',            # directory for storing logs
        logging_steps=10,
        eval_strategy="epoch",            # Evaluate every epoch
        save_strategy="epoch",           # Save checkpoints every epoch
        load_best_model_at_end=True,     # Load the best model at the end of training
        metric_for_best_model="accuracy", # Use accuracy to determine the best model
        greater_is_better=True,           # Higher accuracy is better
        report_to="wandb"                # Report metrics and logs to Weights & Biases
    )

    # Log hyperparameters to W&B
    wandb.config.update(training_args.to_dict())

    # Create Trainer instance with W&B integrated training arguments
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=test_dataset,
        compute_metrics=compute_metrics
    )

    # Train the model (training metrics will be logged automatically)
    print("\nTraining the model with W&B tracking...")
    trainer.train()

    # Evaluate the model (evaluation metrics will be logged automatically)
    print("\nEvaluating the model with W&B tracking...")
    eval_results = trainer.evaluate()
    print("\nEvaluation results:")
    display(eval_results)

    # Finish the W&B run
    wandb.finish()

    print("W&B run finished.")

else:
    print("DataFrame is empty after loading and preprocessing. Cannot proceed with training.")
```
